# Standardization : 
 -> It is a very effective technique which re-scales a feature value so that it has distribution with 0 mean value and variance equals to 1.

 #  Xnew = (Xi - X`mean) / Standard Deviation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset  = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(), inplace=True)
```

```python
sns.displot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["X"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.displot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.displot(dataset["ApplicantIncome_ss"])
plt.show()
```
