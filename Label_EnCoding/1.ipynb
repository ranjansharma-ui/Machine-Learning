```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["cat","cow","dog","black"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit_transform(df["name"])
```

```python
df["en_data"] = le.fit_transform(df["name"])
df
```
