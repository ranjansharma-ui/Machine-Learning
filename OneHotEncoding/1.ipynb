```python
import pandas as pd
```

```python
detaset = pd.read_csv("loan.csv")
detaset.head(5)
```

```python
detaset.isnull().sum()
```

```python
detaset["Gender"].fillna(detaset["Gender"].mode()[0],inplace=True)
```

```python
detaset["Married"].fillna(detaset["Married"].mode()[0],inplace=True)
```

```python
en_data = detaset[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(en_data) # data trinsform  categorical to numerical
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_No","Married_Yes"])
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male"])
```
